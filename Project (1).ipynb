```python
import pandas as pd
df = pd.read_csv("lahore_aqi_2019_to_2023.csv")
df.head()
```

```python
# in which  we have to predict aqi_pm2.5  which is an target col
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
df.info()
```

```python
df=df.sort_values("date")
```

```python
df.set_index('date', inplace=True)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# filling missing values with mean
df["aqi_pm2.5"].fillna(df["aqi_pm2.5"].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
df.info()   #now aqi values are also 1644, values filled with mean
```

```python
df.head()
```

```python
df['aqi_lag_1'] = df['aqi_pm2.5'].shift(1)
df['aqi_lag_2'] = df['aqi_pm2.5'].shift(2)
df['aqi_lag_7'] = df['aqi_pm2.5'].shift(7)
#lag transformations
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df = df.drop([
    "max_temp_f",
    "max_dew_point_f",
    "max_humidity_percent",
    "max_wind_speed_mph",
    "max_pressure_in"
], axis=1)
```

```python
df = df.drop([
    "min_temp_f",
    "min_dew_point_f",
    "min_humidity_percent",
    "min_wind_speed_mph",
    "min_pressure_in"
], axis=1)
```

```python
X = df.drop("aqi_pm2.5", axis=1)         #axis=1 means we are talking about columns
y = df["aqi_pm2.5"]
```

```python
X            #features
```

```python
y       # to predict
```

```python
train_size = int(len(df) * 0.8)
X_train = X.iloc[:train_size]
X_test = X.iloc[train_size:]

y_train = y.iloc[:train_size]
y_test = y.iloc[train_size:]
```

```python
train_size
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("R Score:", r2)
print("Error:", mae)
print("MSE:", mse)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# If not already done:
# df = your cleaned dataframe after creating lags and dropping NaNs

plt.figure(figsize=(10, 8))
corr = df.corr()               # correlation matrix of all numeric columns

sns.heatmap(corr,
            annot=True,        # show numbers
            fmt=".2f",         # format to 2 decimal places
            square=True)

plt.title("Correlation Heatmap of AQI and Features")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         linestyle="--")
plt.xlabel("Actual AQI")
plt.ylabel("Predicted AQI")
plt.title("Actual vs Predicted AQI")
plt.show()


# The model captures typical AQI levels reasonably well, but it underestimates very high pollution episodes
# This is likely because extreme AQI spikes depend on additional factors such as traffic intensity, industrial emissions, construction dust,
# or burning activity, which are not included as features in our dataset. As a result, the model cannot fully explain these sudden peaks.
```

```python
plt.figure(figsize=(10, 4))
plt.plot(y_test.index, y_test, label="Actual AQI")
plt.plot(y_test.index, y_pred, label="Predicted AQI")
plt.xlabel("Time")
plt.ylabel("AQI")
plt.title("Actual vs Predicted AQI Over Time (Test Set)")
plt.legend()
plt.show()

#The model follows the general trend of AQI but fails to capture the sharp spikes.
#This happens because many causes of extreme pollution episodessuch as traffic congestion, industrial emissions, burning activity,
# and construction dustare not included as input features, so the model has no information about them.
# In addition, we use a simple linear regression model, which tends to produce smooth, average-like predictions and is less able to represent sudden, non-linear jumps.
```
