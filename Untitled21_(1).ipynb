```python
import pandas as pd
df = pd.read_csv("lahore_aqi_2019_to_2023.csv")
df.head()
```

```python
# in which  we have to predict aqi_pm2.5  which is an target col
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
df.info()
```

```python
df=df.sort_values("date")
```

```python
df.set_index('date', inplace=True)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# filling missing values with mean
df["aqi_pm2.5"].fillna(df["aqi_pm2.5"].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df['aqi_lag_1'] = df['aqi_pm2.5'].shift(1)
df['aqi_lag_2'] = df['aqi_pm2.5'].shift(2)
df['aqi_lag_7'] = df['aqi_pm2.5'].shift(7)
#lag transformations
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
X = df.drop("aqi_pm2.5", axis=1)
y = df["aqi_pm2.5"]
```

```python
X
```

```python
y
```

```python
train_size = int(len(df) * 0.8)
X_train = X.iloc[:train_size]
X_test = X.iloc[train_size:]

y_train = y.iloc[:train_size]
y_test = y.iloc[train_size:]
```

```python
train_size
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("R Score:", r2)
print("Error:", mae)
```
